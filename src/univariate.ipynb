```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a visually appealing style for the plots
sns.set_style("whitegrid")

# Load your dataset from the CSV file
# Make sure the CSV file is in the same directory as your script, or provide the full path.
steam_df = pd.read_csv('../data/updated_steam_games.csv')


```

```python
# --- 1. Descriptive Statistics ---
# Use the .describe() method to get a summary of the dlc_count column
print("--- Descriptive Statistics for DLC Count ---")
print(steam_df['dlc_count'].describe())
print("\n" + "="*40 + "\n")


# --- 2. Histogram Visualization ---
# Create a histogram to see the frequency of different DLC counts.
# A log scale on the y-axis is essential here to see the distribution
# for games with more than zero DLCs.
plt.figure(figsize=(10, 6))
sns.histplot(steam_df['dlc_count'], bins=50, kde=False, color='skyblue')
plt.title('Distribution of DLC Count (Log Scale for Y-Axis)')
plt.xlabel('Number of DLCs')
plt.ylabel('Number of Games (Log Scale)')
plt.yscale('log')
plt.tight_layout()
plt.show()


# --- 3. Box Plot Visualization ---
# A box plot is great for clearly showing the median, quartiles,
# and identifying the range of outliers.
plt.figure(figsize=(8, 6))
sns.boxplot(y=steam_df['dlc_count'], color='lightgreen')
plt.title('Box Plot of DLC Count')
plt.ylabel('Number of DLCs')
plt.yscale('log') # A log scale helps visualize the distribution of the outliers
plt.tight_layout()
plt.show()
```

```python
print("--- Descriptive Statistics for Peak CCU ---")
print(steam_df['peak_ccu'].describe())
print("\n" + "="*40 + "\n")

plt.figure(figsize=(10, 6))
sns.histplot(steam_df['peak_ccu'], bins=50, kde=False, color='skyblue')
plt.title('Distribution of Peak CCU (Log Scale for Y-Axis)')
plt.xlabel('Peak CCU')
plt.ylabel('Number of Games (Log Scale)')
plt.yscale('log')
plt.tight_layout()
plt.show()
```
